```python
grocery_items = []

def add_item():
    name = input("Enter item name: ")
    quantity = input("Enter item quantity: ")
    price = input("Enter item price: ")
    grocery_items.append({"name": name, "quantity": quantity, "price": price})
    print("Item added successfully!")

def display_items():
    print("Grocery Items:")
    if not grocery_items:
        print("No items found.")
    else:
        for item in grocery_items:
            print(f"Name: {item['name']}, Quantity: {item['quantity']}, Price: {item['price']}")
        print()

def search_item():
    name = input("Enter item name to search: ")
    found = False
    for item in grocery_items:
        if item['name'] == name:
            print(f"Item found - Name: {item['name']}, Quantity: {item['quantity']}, Price: {item['price']}")
            found = True
            break
    if not found:
        print("Item not found.")

def categorize_strings(data):
    list_1 = []
    list_2 = []
    list_3 = []

    for string in data:
        length = len(string)
        if 5 < length < 15:
            list_1.append(string)
        elif length > 15:
            list_2.append(string)
        else:
            list_3.append(string)

    return list_1, list_2, list_3

def menu():
    print("Welcome to the Grocery Items Program!")
    while True:
        print("Menu:")
        print("1. Add Item")
        print("2. Display Items")
        print("3. Search Item")
        print("4. Categorize Strings")
        print("5. Exit")

        choice = input("Enter your choice: ")
        print()

        if choice == '1':
            add_item()
        elif choice == '2':
            display_items()
        elif choice == '3':
            search_item()
        elif choice == '4':
            data = ["The living light", "Happy", "Brave", "Old boy", "Simple", "attenuate", " ", "Shy Ones",
                    "Toxic things", "Silent words", "Big Daddy", "Juvenile", "Mothers love", "the foxes",
                    "Nigeria Election", "IT today"]
            result = categorize_strings(data)
            print("List 1:", result[0])
            print("List 2:", result[1])
            print("List 3:", result[2])
            print()
        elif choice == '5':
            print("Exiting the program...")
            break
        else:
            print("Invalid choice. Please try again.")
        print()

# Run the program
menu()
```
