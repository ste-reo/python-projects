```python
grocery_items = []

def add_item():
    item_name = input("Enter item name: ")
    item_quantity = input("Enter item quantity: ")
    item_price = input("Enter item price: ")
    
    item = {
        'name': item_name,
        'quantity': item_quantity,
        'price': item_price,
    }
    grocery_items.append(item)
    print("Item added successfully.")

def display_items():
    if not grocery_items:
        print("No items in the list.")
    else:
        print("Grocery Items:")
        for item in grocery_items:
            print(f"Name: {item['name']}, Quantity: {item['quantity']}, Price: {item['price']}")
    
def search_item():
    search_name = input("Enter item name to search: ")
    
    found_items = []
    for item in grocery_items:
        if item['name'] == search_name:
            found_items.append(item)
    
    if not found_items:
        print("Item not found.")
    else:
        print("Search Results:")
        for item in found_items:
            print(f"Name: {item['name']}, Quantity: {item['quantity']}, Price: {item['price']}")
    
def display_menu():
    print("\nMenu:")
    print("1. Entry Module")
    print("2. Exit Module")
    print("3. Display Module")
    print("4. Search Module")
    print("0. Exit Program")

def grocery_program():
    while True:
        display_menu()
        choice = input("Enter your choice: ")
        
        if choice == '1':
            add_item()
        elif choice == '2':
            print("Exiting the program...")
            break
        elif choice == '3':
            display_items()
        elif choice == '4':
            search_item()
        elif choice == '0':
            print("Exiting the program...")
            break
        else:
            print("Invalid choice. Please try again.")

# Example usage
grocery_program()
```
